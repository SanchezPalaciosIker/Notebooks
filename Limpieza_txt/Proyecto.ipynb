### a) Lectura del archivo *Nombres.txt*

```python
import pandas as pd

data = pd.read_csv('Nombres.txt', sep="|")
data.columns
data.head()
```

### b) Limpieza de los campos Nombre y Apellido

```python
# Nombre
data['Nombre'] = data['Nombre'].str.replace(',', '') # Quitamos comas
data['Nombre'] = data['Nombre'].apply(lambda x : unidecode(x)).str.strip() # Quitamos tildes y recortamos espacios
data['Nombre'] = data['Nombre'].str.title() # Definimos iniciando con mayúsculas


# Apellido
from unidecode import unidecode # Ayudará a quitar tildes
data['Apellido'] = data['Apellido'].str.replace('@', '').str.replace('_', '') # Quitamos @'s y _'s 
data['Apellido'] = data['Apellido'].apply(lambda x : unidecode(x)).str.strip() # Quitamos tildes y recortamos espacios
data['Apellido'] = data['Apellido'].str.title() # Definimos iniciando con mayúsculas


# Saldo
data['Saldo'] = data['Saldo'].apply(lambda x: "{:.2f}".format(x)) #Quitamos notación científica
data['Saldo'] = pd.to_numeric(data['Saldo'])


data.head()
```

### c) Crea un campo que se llame First que contenga la primera letra del apellido

```python
data['First'] = data['Apellido'].apply(lambda x : x[0:1])
data.head()
```

### d) Usando el campo First, filtra las personas que su nombre inicia con consonante

```python
consonantes = ['B', 'C', 'D', 'F', 'G', 'H', 'J', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'X', 'Y', 'Z']
data[data['First'].isin(consonantes)].head() # Consulta limitada a 5 registros
```

### e) Agrupación de la tabla anterior por los campos Nombre, Apellido, Id con suma del saldo.

```python
agrupacion = data.groupby(by = ['Nombre', 'Apellido', 'Id'])['Saldo'].sum().reset_index()
agrupacion.head() #Agrupación limitada a 5 registros
```

### f) Exportación de la tabla agrupada como archivo Excel con el nombre Nombres_agrupado.xlsx

```python
nombre_archivo = 'Nombres_agrupado.xlsx'
hoja_trabajo = 'Sheet1'
agrupacion.to_excel(nombre_archivo, sheet_name = hoja_trabajo, index=False)
```
